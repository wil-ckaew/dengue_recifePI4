```python
df.loc[ df['ms_notificacao'] == 'Jan', 'ms_notificacao'] = 1
df.loc[ df['ms_notificacao'] == 'Fev', 'ms_notificacao'] = 2
df.loc[ df['ms_notificacao'] == 'Mar', 'ms_notificacao'] = 3
df.loc[ df['ms_notificacao'] == 'Abr', 'ms_notificacao'] = 4
df.loc[ df['ms_notificacao'] == 'Mai', 'ms_notificacao'] = 5
df.loc[ df['ms_notificacao'] == 'Jun', 'ms_notificacao'] = 6
df.loc[ df['ms_notificacao'] == 'Jul', 'ms_notificacao'] = 7
df.loc[ df['ms_notificacao'] == 'Ago', 'ms_notificacao'] = 8
df.loc[ df['ms_notificacao'] == 'Set', 'ms_notificacao'] = 9
df.loc[ df['ms_notificacao'] == 'Out', 'ms_notificacao'] = 10
df.loc[ df['ms_notificacao'] == 'Nov', 'ms_notificacao'] = 11
df.loc[ df['ms_notificacao'] == 'Dez', 'ms_notificacao'] = 12
```
